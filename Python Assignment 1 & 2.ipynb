# 2.	Write a Python program which accepts the user's first and last name and print them in reverse order with a space between them.

```python
fname = input("Input your First Name: ")
lname = input("Input your Last Name: ")

add = fname + " " + lname

list_1 = add.split()
print(add)

reverse_String = ""  # Empty String  
count = len(add) # Find length of a string and save in count variable  
while count > 0:   
    reverse_String += add # save the value of str[count-1] in reverseString  
    count = count - 1 # decrement index  
print ("The reversed string using a while loop is : ",reverse_String)# reversed string
 
```

# 4.	Write a Python program to count the number of characters (character frequency) in a string.

```python
def char_frequency(str1):
    dict = {}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
print(char_frequency('google.com'))
```

# 5.	Write a Python program to get a single string from two given strings, separated by a space and swap the first two characters of each string.

```python
def chars_mix_up(a, b):
  new_a = b[:2] + a[2:]
  new_b = a[:2] + b[2:]

  return new_a + ' ' + new_b
print(chars_mix_up('abc', 'xyz'))
```

# 7.	Write a Python program to sum all the items in a list.

```python
def sum_list(items):
    sum_numbers = 0
    for x in items:
        sum_numbers += x
    return sum_numbers
print(sum_list([1,2,-8]))
```

# 17.	Write a python program to find whether a given year is a leap year or not

```python
year = int(input("Enter Year: "))

if year % 4 == 0 and year % 100 != 0:
    print(year, "is a Leap Year")
elif year % 100 == 0:
    print(year, "is not a Leap Year")
elif year % 400 ==0:
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
```

# 20.	Write a python program to check whether a triangle is Equilateral, Isosceles or Scalene 

```python
print("Input lengths of the triangle sides: ")
x = int(input("x: "))
y = int(input("y: "))
z = int(input("z: "))

if x == y == z:
	print("This is an Equilateral triangle")
elif x==y or y==z or z==x:
	print("This is an isosceles triangle")
else:
	print("This is a Scalene triangle")
```
